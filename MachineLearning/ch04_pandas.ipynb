```python
import pandas as pd
raw_data = {
'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
'test_score': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}

df_left = pd.DataFrame(raw_data, columns = ['subject_id', 'test_score'])
df_left
```

```python
raw_data = {
            'subject_id': ['4', '5', '6', '7', '8'],
            'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
            'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}

df_right = pd.DataFrame(raw_data, columns = ['subject_id', 'first_name', 'last_name'])
df_right
```

```python
pd.merge(left=df_left, right=df_right, how="inner", on='subject_id')
```

```python
pd.merge(df_left, df_right, on='subject_id', how='outer')
```

```python
import os
path = "./data"
filenames = [os.path.join(path, filename)
           for filename in os.listdir(path) if "sales" in filename]
print(filenames)#파일 이름
```

```python
#filenames 기반으로 read_csv
df_list = [pd.read_csv(f) for f in filenames]

for df in df_list:
   print(type(df), len(df))
```

```python
df = pd.concat(df_list, axis=0) #axis=0으로 rk로로 연결
df.reset_index(drop=True)#중복된 인덱스를 제거 및 인덱싱 다시하기
print(len(df))
```
