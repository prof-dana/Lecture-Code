```python
import pandas as pd# (1) pandas 모듈 호출

#(2)데이터 URL을 변수 data_url에 넣기
data_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data'

#(3) csv 데이터 로드
df_data = pd.read_csv(data_url, sep='\s+', header = None)

#(4) 데이터의 열 이름 지정
df_data.columns = ['CRIM','ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO' ,'B', 'LSTAT', 'MEDV']
df_data.head()
```

```python
df_data['weight_0'] = 1 # weight 0 값 추가
df_data.head()
```

```python
y = df_data.pop("MEDV")# Y 추출
x = df_data.values #x 값 행렬(Matrix) 데이터로 변환하기
print(x)
```

```python
import numpy as np#numpy 모듈 호출
weight_vector = np.random.choice(10, 14, replace=True)#가중치 w 생성
y_hat=x.dot(weight_vector) #내적 연산 실행 결과=>예상값
y_hat
```
